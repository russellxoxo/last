```
import string
import random
import re

def RandomSymbols(length):
  lettersAndDigits = [] 
  for sym in (string.ascii_letters + string.digits):
    lettersAndDigits.append(sym)
  randomValues = []
  for i in range(0, length):
    randomValues.append(random.choice(lettersAndDigits))
  return randomValues

poem = "Распустился ландыш в мае, в самый праздник — в первый день. Май цветами провожая, распускается сирень."
poemList = re.findall(r"[\w']+", poem)
for elem in poemList:
  if elem == '': 
    poemList.remove(elem)
print(poemList)
randomValues = RandomSymbols(len(poemList))
hashTable = dict(zip(poemList, randomValues))
for i in hashTable:
  print(i, " - ", hashTable[i])
```
